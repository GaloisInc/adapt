```python
def bagofwords(s,sep):
    return {w for w in s.split(sep) if w != ''}
bagofwords(" a b cde f ",' ')
```

```python
def allbagofwords(s,sep):
    bags = set()
    for x in s:
        bags = bags | bagofwords(x,sep)
    return bags

allbagofwords({'a b cde f', ' gh i jkl '}, ' ')
```

```python
context3 = {k:allbagofwords(context[k],'/') for k in context.keys()}
dictionary3 = {}
cost(context3,dictionary3)
```

```python
rePair(context3,dictionary3,thresh=2,showCost=True)
```
