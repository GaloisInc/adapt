```python
import random

def generate_node_attributes():
    return {'normalcy': random.random(), 'labels': [('abc', random.random()), ('def', random.random())]}

def random_graph(n, p, seed=None):
    G=nx.DiGraph()
    
    for i in xrange(n):
        node_attrs = generate_node_attributes()
        G.add_node(i, attr_dict=node_attrs)
    
    for i in xrange(n):
        for j in xrange(i):
             if random.random() < p:
                G.add_edge(i,j)
    
    return G

G = random_graph(20, 0.1)
pos=nx.shell_layout(G)

nx.draw(G, pos=pos)
nx.draw_networkx_labels(G, pos=pos, labels={x:x for x in G.nodes()})

print G.nodes(data=True)
print G.nodes()
    
```
